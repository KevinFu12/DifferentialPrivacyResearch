```python
import numpy as np

population_size = 10
num_generations = 10
crossover_rate = 0.8
mutation_rate = 0.02
num_features = X.shape[1]
```

```python
def initialize_population():
    return [np.random.choice([0,1], size=(num_features)) for _ in range(population_size)]
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.metrics import accuracy_score, accuracy_score, f1_score, roc_auc_score, precision_score, recall_score

def fitness_function(chromosome, X, y):
    selected_indices = np.where(chromosome == 1)[0]

    if len(selected_indices) == 0:
        return 0

    X_selected = X[:, selected_indices]

    XTrain, XTest, yTrain, yTest = train_test_split(X_selected, y, test_size=0.2, random_state=42)

    rf = RandomForestClassifier(n_estimators=50, class_weight="balanced", random_state=42)
    rf.fit(XTrain, yTrain)

    yPred = rf.predict(XTest)
    f1 = f1_score(yTest, yPred)
    acc = accuracy_score(yTest, yPred)

    return f1
```

```python
import random

def select_parents(population, fitness_scores):
    parents = []
    for _ in range(2):
        tournament = random.sample(list(zip(population, fitness_scores)), 3)
        best = max(tournament, key=lambda x: x[1])
        parents.append(best[0])
    return parents
```

```python
def crossover(parent1, parent2):
    if random.random() < crossover_rate:
        point = random.randint(1, num_features - 1)
        child1 = np.concatenate([parent1[:point], parent2[point:]])
        child2 = np.concatenate([parent2[:point], parent1[point:]])
    else:
        child1, child2 = parent1.copy(), parent2.copy()

    return child1, child2
```

```python
population = initialize_population()

best_solution = None
best_fitness = 0

for generation in range(num_generations):
    print(f"Generation {generation+1}")

    fitness_scores = [fitness_function(ind, X, y) for ind in population]

    max_fitness = max(fitness_scores)
    max_index = fitness_scores.index(max_fitness)

    if max_fitness > best_fitness:
        best_fitness = max_fitness
        best_solution = population[max_index]

    print(f"  Best F1-score so far: {best_fitness:.4f}")

    new_population = []

    while len(new_population) < population_size:
        parent1, parent2 = select_parents(population, fitness_scores)
        child1, child2 = crossover(parent1, parent2)
        child1 = mutate(child1)
        child2 = mutate(child2)
        new_population.extend([child1, child2])

    population = new_population[:population_size]

print("\nGA Completed!")
print(f"Best Solution (Selected Features): {best_solution}")
print(f"Best F1-score Achieved: {best_fitness:.4f}")
```

```python
selected_indices = np.where(best_solution == 1)[0]
selected_feature_names = X_df.columns[selected_indices].tolist()

print("\n Selected Features:", selected_feature_names)

X_selected = X[:, selected_indices]

X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, stratify=y, random_state=42)

final_rf = RandomForestClassifier(n_estimators=100, random_state=42)
final_rf.fit(X_train, y_train)

y_pred = final_rf.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
auc = roc_auc_score(y_test, y_pred)

print("\n Final Model Evaluation:")
print(f"  Accuracy  : {accuracy:.4f}")
print(f"  Precision : {precision:.4f}")
print(f"  Recall    : {recall:.4f}")
print(f"  F1-score  : {f1:.4f}")
print(f"  AUC       : {auc:.4f}")
```
